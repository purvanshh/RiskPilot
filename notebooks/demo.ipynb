# Loan Approval Multi-Agent Underwriter Demo

This notebook provides a simple demonstration of running the LangGraph underwriting pipeline.

```python
import sys
sys.path.append('..')

import json
from src.graph.state import LoanApplicationState, ExtractedDocument
from src.graph.graph import graph

# Load sample applications
with open('../data/test_applications.json', 'r') as f:
    apps = json.load(f)

target_app = apps[0]
print(f"Loaded Application: {target_app['application_id']} for {target_app['applicant_data']['name']}")
```

```python
# Setup state
docs = [ExtractedDocument(**d) for d in target_app['documents']]
state = LoanApplicationState(
    application_id=target_app['application_id'],
    applicant_data=target_app['applicant_data'],
    documents=docs
)

# Run the graph pipeline
result = graph.invoke(state)
print("Recommendation:", result['arbitrator_output'].recommendation)
print("Summary:", result['arbitrator_output'].summary)
```
